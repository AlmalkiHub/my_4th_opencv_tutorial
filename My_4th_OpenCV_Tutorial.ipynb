## **Drawing images and shapes using OpenCV**



````python
# Import required libraries 
import cv2
import numpy as np 
import matplotlib.pyplot as plt 
````

````python
# Create a helper function to display an image 
def display_image(title="", image=None, size=10):
  """
  This function takes an image as an input and returns a displayed image.
  """

  w, h = image.shape[0], image.shape[1]
  aspect_ratio = w/h 
  plt.figure(figsize=(size * aspect_ratio, size))
  plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
  plt.title(title)
  plt.show() # display the image
````

#### Let's start off by making a blank square canvas


````python
# Create a black image using numpy an array of black 
image = np.zeros((512,512,3), np.uint8)
image.shape
````

````python
# Make this in black and white (grayscale)
image_grayscale = np.zeros((512, 512), np.uint8)
image_grayscale.shape
````

````python
# Plot the two images 
display_image("Black Canvas - RGB Color Channels", image)
display_image("Black Canvas - Grayscale", image_grayscale)
````

#### **Let's draw a line over our black square**

```cv2.line(image, starting cordinates, ending cordinates, color, thickness)```

````python
# Draw a line over black suqare image (it changes the input image)
cv2.line(image, (0,0), (511, 511), (255, 127, 0), 5)
````

````python
# Display the image again 
display_image("Black Canvas with Diagonal Line", image)
````

#### **Drawing Rectangles**

```cv2.rectangle(image, starting vertex, opposite vertex, color, thickness)```

````python
# Create our black canvas again because now it has a line in it
image = np.zeros((512,512,3), np.uint8)
image.shape
````

````python
# Thickness - if positive. Negative thickness means that it is filled
cv2.rectangle(image, (200,200), (300,250), (127,50,127), 10)
display_image("Black Canvas With Pink Rectangle", image)
````

#### **Drawing Cirlcles**

```cv2.cirlce(image, center, radius, color, fill)```

````python
# Draw the image again because it has a rectagnle in it 
image = np.zeros((512,512,3), np.uint8)
image.shape
````

````python
cv2.circle(image, (300, 300), 100, (255,255,255), -1) 
display_image("Black Canvas With Green Circle", image)
````

### **Drawing Polygons**

```cv2.polylines(image, points, Closed?, color, thickness)```

if Closed = True, we join the first and last points.

````python
# Create the image again
image = np.zeros((512,512,3), np.uint8)
image.shape
````

````python
# Let's define four points
pts = np.array( [[10,50], [400,50], [90,200], [50,500]], np.int32)

# Let's now reshape our points in form  required by polylines
pts = pts.reshape((-1,1,2))

cv2.polylines(image, [pts], True, (0,0,255), 3)
display_image("Black Canvas with Red Polygon", image)
````

### **Adding text with cv2.putText**

```cv2.putText(image, 'Text to Display', bottom left starting point, Font, Font Size, Color, Thickness)```

**Available Fonts**
- FONT_HERSHEY_SIMPLEX, FONT_HERSHEY_PLAIN
- FONT_HERSHEY_DUPLEX,FONT_HERSHEY_COMPLEX 
- FONT_HERSHEY_TRIPLEX, FONT_HERSHEY_COMPLEX_SMALL
- FONT_HERSHEY_SCRIPT_SIMPLEX
- FONT_HERSHEY_SCRIPT_COMPLEX

````python
# Create the image 
image = np.zeros((2000,2000,3), np.uint8)
````

````python
# Define the text to be added
text_to_display =  'Ali Almalki. Best Data Scientist in the world!!'
# Plot the image with the text
cv2.putText(image, text_to_display, (155,290), cv2.FONT_HERSHEY_PLAIN, 3, (125,200,125), 5)
display_image("Adding text to the image", image)
````
